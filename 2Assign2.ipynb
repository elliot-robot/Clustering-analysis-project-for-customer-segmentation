#Task-1

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

df=data

# 1. Handling missing values
# Checking for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Re-importing necessary libraries and variables
import pandas as pd
from sklearn.preprocessing import StandardScaler


# One-hot encoding for 'Channel' and 'Region'
data_encoded = pd.get_dummies(df, columns=['Channel', 'Region'], drop_first=True)

# Standardizing the numerical features
numerical_columns = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Delicassen']
scaler = StandardScaler()
data_encoded[numerical_columns] = scaler.fit_transform(data_encoded[numerical_columns])

# Converting boolean True/False to 1/0 in the one-hot encoded columns
data_encoded[['Channel_2', 'Region_2', 'Region_3']] = data_encoded[['Channel_2', 'Region_2', 'Region_3']].astype(int)

# Displaying the updated data to confirm the change
data_encoded.head()
```

```python
data_encoded.to_csv('data_after_preprocessing.csv',index=False)
```

```python
df1=data_encoded
```

```python
# Ensuring df1 contains only numeric columns
df1 = df1.select_dtypes(include=[float, int])
df1 = df1.sample(frac=0.1)  
df1
```

```python
#Task-2
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df1)  # Ensure df1 contains only numeric data
    sse.append(kmeans.inertia_)

# Plotting the elbow curve
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Performing hierarchical clustering
linked = linkage(data_encoded, 'ward')

# Create a dendrogram to visualize the clustering
plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Load preprocessed data
df1 = pd.read_csv('data_after_preprocessing.csv')  # replace with the correct path

# Assuming we already have the optimal number of clusters from Task 2
optimal_k = 3  # Adjust based on elbow method from Task 2
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(df1)

# Getting cluster centroids
centroids = kmeans.cluster_centers_
centroids_df = pd.DataFrame(centroids, columns=df1.columns)

# Displaying centroids
print("Cluster centroids:\n", centroids_df)
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering

# Silhouette Score for K-Means
kmeans_labels = kmeans.labels_
kmeans_silhouette = silhouette_score(df1, kmeans_labels)
print(f'Silhouette Score for K-Means: {kmeans_silhouette:.4f}')

# Silhouette Score for Hierarchical Clustering
hier_cluster = AgglomerativeClustering(n_clusters=optimal_k, linkage='ward')
hier_labels = hier_cluster.fit_predict(df1)
hier_silhouette = silhouette_score(df1, hier_labels)
print(f'Silhouette Score for Hierarchical Clustering: {hier_silhouette:.4f}')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Loading the data 
data = df1

# Converting one-hot encoded Region columns to a single label
data['Region'] = data[['Region_2', 'Region_3']].idxmax(axis=1).apply(lambda x: int(x.split('_')[-1]))

# Splitting the features and target
X = data.drop(columns=['Region', 'Region_2', 'Region_3'])
y = data['Region']

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initializing and train the SVM model
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)

# Predictions and evaluations
y_pred = svm_model.predict(X_test)
print("SVM Classification Report:\n", classification_report(y_test, y_pred))
```
